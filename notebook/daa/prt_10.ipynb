```python
bs_dt = "20260108"
output_dir = Path(f"../../backtest/daa_backtest_outputs/{bs_dt}")
pattern = "*.parquet"
files = sorted(output_dir.glob(pattern))
files
```

```python
res_df = pd.DataFrame()

res_files = sorted(
    f for f in output_dir.glob(pattern)
    if "res" in f.name
)

for f in res_files:
    df = pd.read_parquet(f)
    res_df = pd.concat([res_df, df])

res_df.sort_values(by=["Sharpe Ratio", "CAGR", "MDD"], ascending=[False, False, False])
```
